```python
from sklearn.linear_model import LinearRegression

X = df.loc[:, ['Age']]
X.dropna(inplace=True)  
y = df.loc[:, 'Quantity'] 
y, X = y.align(X, join='inner')  #

model = LinearRegression()
model.fit(X, y)

y_pred = pd.Series(model.predict(X), index=X.index)
```
